```python
#Homework Problem 10_Sahana Bhattacharyya
#Finding Binding Energy

import numpy as np

A=float(input("Mass Number "))
Z=float(input("Atomic Number "))

b=15.67
c=17.23
d=0.75
e=93.2

if A%2==0 and Z%2==0:
    f=12.0
elif A%2==0 and Z%2==1:
    f=-12.0
elif A%2==1:
    f=0
    
p=b*A
q=c*(A**(2/3))
r=d*(Z**2)/(A**(1/3))
s=e*((A-2*Z)**2)/A
t=f/(A**(1/2))

B=p-q-r-s+t
    
print("Binding Energy",B)
```

```python
#Binding Energy per nucleon for A =(Z,3*Z) and finding the maximum Binding Energy per Nucleon Value
import numpy as np

Z=int(input("Atomic Number "))
A = np.arange(Z,3*Z+1)
print(A)

b=15.67
c=17.23
d=0.75
e=93.2

size= 2*Z+1
p,q,r,s,t,B = [np.zeros(shape=(1,size))]*6

for i in range (size):
    if (A[i]%2)==0 and Z%2==0:
        f=12.0
    elif A[i]%2==0 and Z%2==1:
        f=-12.0
    elif A[i]%2==1:
        f=0

p=b*A
q=c*(A**(2/3))
r=d*(Z**2)/(A**(1/3))
s=e*((A-2*Z)**2)/A
t=f/(A**(1/2))

B=p-q-r-s+t
    
print("Binding Energy per nucleon",B/A)

K=(B/A).max()
L=np.array(B/A)
np.where(L == (B/A).max())

pos = L.argmax()                      #"L.argmax()" gives the position of the highest value in an array
print("The value of A for which Binding Energy per Nucleon is highest is ", A[pos])

print("Maximum Binding Energy per Nucleon is ", K)
```
